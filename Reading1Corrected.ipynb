**1. Why the error affecting numerical schemes, such as schemes for numerically solving motion of gravitationally interacting particles, is called truncation error? Check all that applies.** 

The correct items to check are: 

* Because numerical schemes neglect (truncates) terms in the Taylor expansion

* Because numerical schemes approximate mathematical derivative with an approximation that involves finite differences

The last item:
"Because numerical calculations are limited by a limited number of bits used to represent floating point numbers"

is not related to truncation error. The limitation due to a finite number of bits is called *roundoff error* and has a different origin (finite number of bits used to represent numbers).

**6. This question is for graduate students only: Why it is advisable to implement equations for numerical computation in dimensionless "code units" rather than in physical units (say, using meters for distance, meters/second for velocity, kg for mass, etc.)? (check all that applies)**

* Code units can be chosen so that computations involve numbers of similar scales minimizing roundoff errors

* Code units can be changed to optimize calculations for very different physical systems without changing modelling code itself

* Code units can be used to streamline equations avoiding unnecessary terms and multiplications
