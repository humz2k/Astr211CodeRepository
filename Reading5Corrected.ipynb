**3. Why in practice we need to sample log of the distribution, not the distribution itself?** 

Answer: This is needed to avoid numerical underflow in areas where distribution amplitude is extremely small

MCMC does not work faster with the logarithm and does not strictly require using logarithm. With some care for certain distributions sampling can be done without using the logarithm. However, in general without using logarithm roundoff underflow errors in the areas of very small values of distribution cause severe problems and often prevent sampling. 

**4. What can we do to minimize the length of the burn-in sequence?** 

Answer: Choose starting location closer to the maximum of the sampled distribution

When we start close to the maximum of the distribution initial samples are much more typical samples of the distribution and so burn-in length is essentially zero.

**5. The trace plots of a variable sampled by MCMC can be used to** 

Answer: 

* To estimate the length of the burn-in sequence

* To roughly estimate the auto-correlation time of the MCMC chain

* To gauge convergence of the MCMC chain

Duplicate samples are not visible in the trace plot so they cannot be identified. 

**6. What is measured by the auto-correlation time?** 

Answer: The typical number of steps over which samples in the chain are significantly correlated

**7. If I have MCMC chain for a parameter x of length 10000 and estimate of the auto-correlation time for this chain is 100. How many truly independent samples do I have in the chain approximately?** 

Answer: The number of independent samples is approximately equal to the length of the chain divided by the auto-correlation time (average width of peaks and valleys in the trace plot): so here it is ~10000/100~100.
